```python
import csv

#define a class Cart - a subclass of list class
class Cart(list):
    #define a subtotal function
    def subtotal(self):
        #creating a value to store the price
        price = 0.0
        for item in self:
            if item.get_price() == 'Free':
                price += 0
            else:
                price += float(item.get_price())
        #Returns subtotal from a Cart list object
        return round(price, 2)
    

class App(object):
    '''App class defines an app item
    available in store. App object saved in
    category_dict per category
    and rating_dict per rating and price_dict per price '''
    category_dict = {}
    rating_dict = {}
    price_dict = {}

    def __init__(self, ID, name, developer, description, price, rating, review_count, category):
        '''Initialization method'''
        self._id = ID
        #initialize name, developer, description, price, rating,
        #review_count, category as private attribute
        self._name = name
        self._developer = developer
        self._description = description
        self._price = price
        self._rating = rating
        self._review_count = review_count
        self._category = category
        #populate dict by category - key = category and values = list of app objects
        if self._category in App.category_dict.keys():
            #append object to current category
            App.category_dict[self._category].append(self)
        else:
            App.category_dict[self._category] = [self]

        #populate rating_dict and price_dict
        #rating_dict key = rating value as whole number
        # for example an app of 4.9 rating will be saved under key 4 and values = list of objects for
        #that rating
        if float(self._rating) >= 4 and float(self._rating) <= 4.9:
            if 4 in App.rating_dict.keys():
                App.rating_dict[4].append(self)
            else:
                App.rating_dict[4] = [self]
        elif float(self._rating) >= 3 and float(self._rating) <= 3.9:
            if 3 in App.rating_dict.keys():
                App.rating_dict[3].append(self)
            else:
                App.rating_dict[3] = [self]
        elif float(self._rating) >= 2 and float(self._rating) <= 2.9:
            if 2 in App.rating_dict.keys():
                App.rating_dict[2].append(self)
            else:
                App.rating_dict[2] = [self]
        elif float(self._rating) >= 1 and float(self._rating) <= 1.9:
            if 1 in App.rating_dict.keys():
                App.rating_dict[1].append(self)
            else:
                App.rating_dict[1] = [self]
        else:
            if 5 in App.rating_dict.keys():
                App.rating_dict[5].append(self)
            else:
                App.rating_dict[5] = [self]
                
        #price_dict will be similar as rating except key will be price,
        #all the apps between 1.00$-1.99 will be under key 1.00$
        if self._price == 'Free' or float(self._price) < 1:
            if '$0.00' in App.price_dict.keys():
                App.price_dict['$0.00'].append(self)
            else:
                App.price_dict['$0.00'] = [self]
        elif float(self._price) >= 5 and float(self._price) < 6:
            if '$5.00' in App.price_dict.keys():
                App.price_dict['$5.00'].append(self)
            else:
                App.price_dict['$5.00'] = [self]
        elif float(self._price) >= 4 and float(self._price) < 5:
            if '$4.00' in App.price_dict.keys():
                App.price_dict['$4.00'].append(self)
            else:
                App.price_dict['$4.00'] = [self]
        elif float(self._price) >= 3 and float(self._price) < 4:
            if '$3.00' in App.price_dict.keys():
                App.price_dict['$3.00'].append(self)
            else:
                App.price_dict['$3.00'] = [self]
        elif float(self._price) >= 2 and float(self._price) < 3:
            if '$2.00' in App.price_dict.keys():
                App.price_dict['$2.00'].append(self)
            else:
                App.price_dict['$2.00'] = [self]
        else:
            if '$1.00' in App.price_dict.keys():
                App.price_dict['$1.00'].append(self)
            else:
                App.price_dict['$1.00'] = [self]
            
    #define get methods
    #to return all private attributes
    def get_id(self):
        return self._id
    
    def get_name(self):
        return self._name
    
    def get_developer(self):
        return self._developer
    
    def get_description(self):
        return self._description
    
    def get_price(self):
        return self._price
    
    def get_rating(self):
        return self._rating
    
    def get_review_count(self):
        return self._review_count
    
    def get_category(self):
        return self._category


#process file
filename = 'Top50ShopifyApps.csv'
with open(filename) as fin:
    apps = csv.reader(fin, delimiter = ',')
    next(apps)
    for row in apps:
        ID, name, developer, description, price, rating, review_count, category = row
        App(ID, name, developer, description, price, rating, review_count, category)
        


'''Testing code to check object creating Items
Uncomment to test and then
Comment out when done'''
'''
for k,v in App.category_dict.items(): #v is a list of all objects
    print(k, [(obj.get_name(), obj.get_rating()) for obj in v ])
print('++++++++++')

for k,v in App.rating_dict.items(): #v is a list of all objects
    print(k, [obj.get_rating() for obj in v ])
print('++++++++++')

for k,v in App.price_dict.items(): #v is a list of all objects
    print(k, [obj.get_price() for obj in v ])
print('++++++++++')
'''






          

```
